# **Test**----> Are a long time between users create their accounts and their first transactions strong churn signals?

```python
import pandas as pd
```

```python
transaction=pd.read_csv('/content/drive/MyDrive/Colab Notebooks/transactions.csv')
transaction
```

```python
churn_condition=(transaction.churn==0)
churn_df=transaction[churn_condition]['days_since_account_creation']
churn_df
```

```python
unchurn_condition=(transaction.churn==1)
unchurn_df=transaction[unchurn_condition]['days_since_account_creation']
unchurn_df
```

```python
import matplotlib.pyplot as plt
from statsmodels.stats.weightstats import ztest

# Perform the z-test
z_score, p_value = ztest(churn_df, unchurn_df)

alpha = 0.05

plt.figure(figsize=(10, 6))

plt.plot(range(len(churn_df)), churn_df, marker='o', linestyle='-', color='blue', label='Churn')

plt.plot(range(len(unchurn_df)), unchurn_df, marker='o', linestyle='-', color='green', label='Unchurn')

if p_value < alpha:
    plt.text(0.5, max(max(churn_df), max(unchurn_df)) + 0.1, 'Reject Null Hypothesis', color='red', fontsize=12, ha='center')
else:
    plt.text(0.5, max(max(churn_df), max(unchurn_df)) + 0.1, 'Fail to Reject Null Hypothesis', color='green', fontsize=12, ha='center')

plt.xlabel('Index')
plt.ylabel('days_since_account_creation')
plt.title('Churn and Unchurn Data with p-value')

plt.legend()

plt.grid(True)
plt.show()

```

```python
from statsmodels.stats.weightstats import ztest
z_score , p_value = ztest(churn_df, unchurn_df)
print(f'p_value: {p_value}')
alpha = 0.05
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis. There is a statistically significant relationship between time taken to make the first transaction and churn.")
else:
    print("Fail to reject the null hypothesis. There is no statistically significant relationship between time taken to make the first transaction and churn.")
```
