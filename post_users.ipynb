```python
# Get the current year
year_now = datetime.now().year
# Calculate the age by subtracting the birth year from the current year
users['age'] = year_now - users['birth_year']
```

```python
# Function to classify age into categories
def classify_age(age):
    if age <= 26:
        return 'Young'
    elif age <= 31:
        return 'Young Adult'
    elif age <= 39:
        return 'Maturity stage'
    elif age <= 59:
        return 'Adult'
    else:
        return 'Senior'
```

```python
# Apply the function to create a new column 'Age_Category'
users['age_category'] = users['age'].apply(classify_age)
users
```

```python
# Calculate value counts for each category
category_counts = users['age_category'].value_counts()

# Calculate total number of entries
total_entries = users['age_category'].count()

# Calculate percentage for each category
percentage_per_category = round((category_counts / total_entries) * 100,2)

print(percentage_per_category)
```

Churn Rates Among Different Age Groups


```python
import pandas as pd
from scipy.stats import chi2_contingency

# Create a contingency table of observed frequencies
observed_frequencies = pd.crosstab(users['age_category'], users['churn'])

# Perform the Chi-squared test
chi2, p_value, _, _ = chi2_contingency(observed_frequencies)

# Interpret the results
alpha = 0.05
if p_value < alpha:
    print("There is a significant difference in churn rates among age groups (reject H0)")
else:
    print("There is no significant difference in churn rates among age groups (fail to reject H0)")
```

```python
import pandas as pd
from datetime import datetime


# Calculate churn status for each user within each quarter
churn_per_age_category = users.groupby(['age_category', 'user_id'])['churn'].max().reset_index()

# Count the number of True and False values for each quarter
churn_counts_per_age_category = churn_per_age_category.groupby('age_category')['churn'].value_counts().unstack().fillna(0)

# Sort the DataFrame based on the index (age_category) in alphabetical order
churn_counts_per_age_category_sorted = churn_counts_per_age_category.sort_index()

churn_counts_df = pd.DataFrame(churn_counts_per_age_category_sorted)

# Print the DataFrame
print(churn_counts_df.columns)
print(churn_counts_df)
```

```python
churn_counts_df = pd.DataFrame(churn_counts_per_age_category_sorted)
churn_counts_df["churn_rates"] = churn_counts_df.iloc[:,1]/(churn_counts_df.iloc[:,1]+churn_counts_df.iloc[:,0])*100
churn_counts_df
```

```python
# Create a DataFrame from churn_counts_per_age_category_sorted
churn_counts_df = pd.DataFrame(churn_counts_per_age_category_sorted)

# Calculate churn rates and add them to the DataFrame
churn_counts_df["churn_rates"] = (churn_counts_df.iloc[:, 1] / (churn_counts_df.iloc[:, 1] + churn_counts_df.iloc[:, 0])) * 100

# Round churn rates to two decimal places
churn_counts_df["churn_rates_round"] = churn_counts_df["churn_rates"].round(2)

# Display the DataFrame
print(churn_counts_df)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Plot
plt.figure(figsize=(10, 6))

# Assuming 'churn_counts_df' is your DataFrame containing churn rates
# Aggregate churn rates by 'Age Category'
churn_by_age_category = churn_counts_df.groupby('age_category')['churn_rates_round'].mean()

# Plot churn rates by age category
churn_by_age_category.plot(kind='bar', color='skyblue')

plt.title('Churn Rates Among Different Age Groups')
plt.xlabel('Age Category')
plt.ylabel('Churn Rates (%)')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
