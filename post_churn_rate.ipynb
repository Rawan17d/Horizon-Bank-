```python
import pandas as pd
from datetime import datetime, timedelta

# Define minimum and maximum dates
min_date = datetime(2018, 1, 1)
max_date = datetime(2019, 5, 16)

# Filter data for transactions within the specified date range
data_within_range = transaction[(transaction['created_date_of_user'] >= min_date) & (transaction['created_date_of_user'] <= max_date)]

# Find the signup month for each user
data_within_range['signup_month'] = data_within_range['created_date_of_user'].dt.to_period('Q')

# Group users into cohorts based on their signup month
cohorts = data_within_range.groupby('signup_month')

# Calculate the number of unique users in each cohort
cohort_sizes = cohorts['user_id'].nunique()
print(cohort_sizes)

# Calculate churned users for each cohort
churned_users = {}
current_date = max_date
for cohort, cohort_data in cohorts:
    last_transaction_dates = cohort_data.groupby('user_id')['created_date_of_transactions'].max()
    churned_users[cohort] = last_transaction_dates[last_transaction_dates < (current_date - timedelta(days=90))].index.tolist()

# Calculate churn rate for each cohort
churn_rates = {}
for cohort, cohort_data in cohorts:
    churned_count = len(set(churned_users.get(cohort,[])) & set(cohort_data['user_id'].unique()))
    total_users = cohort_sizes.loc[cohort]
    churn_rates[cohort] = (churned_count / total_users) *100

# Print cohort analysis results
print("Cohort\t\t\t\tChurn Rate (%)")
for cohort, churn_rate in churn_rates.items():
    print(f"{cohort}\t\t\t{round(churn_rate,2)}")
```

```python
import pandas as pd
from scipy.stats import chi2_contingency

# Create a contingency table of observed frequencies
observed_frequencies = pd.crosstab(users['age_category'], users['churn'])

# Perform the Chi-squared test
chi2, p_value, _, _ = chi2_contingency(observed_frequencies)

# Interpret the results
alpha = 0.05
if p_value < alpha:
    print("There is a significant difference in churn rates among age groups (reject H0)")
else:
    print("There is no significant difference in churn rates among age groups (fail to reject H0)")
```
